6. Write a function definition named mode that takes in sequence of numbers and
returns the most commonly occuring value

````python
def mode(numbers_sequence):

    if not numbers_sequence:
        return None

    # Use a dictionary to store the frequency of each number
    frequency_map = {}
    for number in numbers_sequence:
        frequency_map[number] = frequency_map.get(number, 0) + 1

    # Find the number with the highest frequency
    most_common_value = None
    max_frequency = 0

    for value, frequency in frequency_map.items():
        if frequency > max_frequency:
            max_frequency = frequency
            most_common_value = value

    return most_common_value

print(mode([1, 2, 2, 3, 4, 2, 5]))
````

Exercise 3: Sort following NumPy array \\
Case 1: Sort array by the second row \\
Case 2: Sort the array by the second column

````python
import numpy as np

sampleArray = np.array ( [ [ 11 ,22, 33 ], [ 44, 55, 66 ], [ 77, 88, 99 ] ] )

# Case 1: Sort array by the second row
second_row = sampleArray[1, :]
sorted_indices_by_second_row = np.argsort(second_row)
sorted_array_by_second_row = sampleArray[:, sorted_indices_by_second_row]

print("\nCase 1: sampleArray sorted by the second row:")
print(sorted_array_by_second_row)

# Case 2: Sort the array by the second column
second_column = sampleArray[:, 1]
sorted_indices_by_second_column = np.argsort(second_column)
sorted_array_by_second_column = sampleArray[sorted_indices_by_second_column, :]

print("\nCase 2: sampleArray sorted by the second column:")
print(sorted_array_by_second_column)

````

Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: â€“

X label name = Month Number \\
Y label name = Total profit \\




````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")


# Extract month number and total profit columns
month_number = df['month_number']
total_profit = df['total_profit']

# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(month_number, total_profit, marker='o', linestyle='-')

# Set plot properties
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.title("Total Profit per Month")
plt.grid(True)
plt.xticks(month_number)
plt.tight_layout()
plt.show()
````

Exercise : Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

# Extract relevant columns
months = df['month_number']
face_cream_sales = df['facecream']
face_wash_sales = df['facewash']

# Plot bar chart
bar_width = 0.4
plt.figure(figsize=(12, 6))

# Bars for face cream
plt.bar(months - 0.2, face_cream_sales, width=bar_width, label='Face Cream', color='skyblue')

# Bars for face wash
plt.bar(months + 0.2, face_wash_sales, width=bar_width, label='Face Wash', color='orange')

# Labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')
plt.xticks(months)  # Set x-axis ticks to month numbers
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
````
